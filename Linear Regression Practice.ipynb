# Linear Regression Practice
A small exercise I did to get more familiar with linear regression analysis in Python. Done using [this dataset](https://www.kaggle.com/datasets/andonians/random-linear-regression/data) on Kaggle. 

### Loading and viewing the data

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy.stats import zscore
import numpy as np
```

```python
# Load the dataset
train_data = pd.read_csv("train.csv")
train_data.head(8).transpose()
```

### Modeling the Data

```python
# Drop rows with missing values
train_data = train_data.dropna()

# Simplify accessing x and y data
x_train = train_data[["x"]]
y_train = train_data["y"]

# Create and fit a LR model
model = LinearRegression()
model.fit(x_train, y_train)
```

### Plotting the training data

```python
# Create scatterplot with regression line
plt.scatter(x_train, y_train, marker="o", label="Training Data")
plt.plot(x_train, model.predict(x_train), color="red", linewidth=3, label="Regression Line")

# Get coefficients and intercept
coefficients = model.coef_
intercept = model.intercept_
equation = "y = " + str(round(coefficients[0], 3)) + "x + " + str(round(intercept, 3))
plt.text(0.1, 100, equation, fontsize=12)

plt.show()
```

### Testing the Model

```python
# Load test data
test_data = pd.read_csv("test.csv")
x_test = test_data[["x"]]
test_data.head().transpose()
```

```python
# Make predictions and add them to the dataframe
y_pred = model.predict(x_test)
model_eval_data = test_data
model_eval_data["y_pred"] = y_pred

# Find residuals
model_eval_data["e"] = model_eval_data["y"] - model_eval_data["y_pred"]
model_eval_data.head()
```

```python
# Create a scatterplot to visualize residuals
plt.scatter(model_eval_data["y_pred"], model_eval_data["e"])
plt.axhline(y=0, color="red", linestyle="--", linewidth=2) # Dashed reference line at y = 0
plt.xlabel("Predicted Values of y")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()
```

The residual plot does not exhibit any clear pattern, so we can conclude that the model does not contain any systematic bias, and is accurate at predicting y given x.
